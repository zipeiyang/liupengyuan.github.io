```python
#第2题另一种思路#
def f1(n):
    sum =1
    for i in range(1,n+1):
        sum*=i
    return sum

def f2(n):
    if n==0:
        return 1
    else:
        return n*f2(n-1)

def main(n):
    sum1=sum2=0
    for i in range(1,n+1):
        sum1+=f1(i)
        
    for i in range(1,n+1):
        sum2+=f2(i)
    print('sum1=%d,sum2=%d'%(sum1,sum2))
    
```
